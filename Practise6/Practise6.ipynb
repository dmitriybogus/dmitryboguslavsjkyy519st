```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

N = 128
t = np.linspace(0, 1, N, endpoint=True)
x = np.sin(2*np.pi*t)
x_rect = np.zeros(N) # прямоугольный сигнал
x_rect[64:96] = 1
plt.plot(x_rect)

plt.figure(figsize = (8, 6))
plt.plot(t, x, 'r')
plt.ylabel('Amplitude')

plt.show()
```

```python
def DFT(x):
    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    
    X = np.dot(e, x) 
    return X
```

```python
def IDFT(x):
    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)

    X = np.dot(e, x)
    Xn = X * 1/N
    return Xn
```

```python
N = len(x)
n = np.arange(N)
freq = n/N
```

```python
X = DFT(x)


plt.figure(figsize = (20, 10))
plt.subplot(121)
plt.stem(freq, np.abs(X), 'b', markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')
plt.ylim(0, 6)

plt.figure(figsize = (20, 10))
plt.subplot(122)
plt.stem(freq, np.angle(X), 'b', markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Phase |X(freq)|')

plt.show()
```

```python
for i in [x, x_rect]:
    X = DFT(i)
    X1 = np.fft.fft(i)

    plt.figure(figsize = (20, 10))
    plt.subplot(121)
    plt.stem(freq, np.abs(X1), 'b', markerfmt=" ", basefmt="-b")
    plt.xlabel('Freq (Hz) np.fft')
    plt.ylabel('DFT Amplitude |X(freq)| np.fft ')
    plt.ylim(0, 6)

    plt.figure(figsize = (20, 10))
    plt.subplot(122)
    plt.stem(freq, np.angle(X1), 'b', markerfmt=" ", basefmt="-b")
    plt.xlabel('Freq (Hz) np.fft')
    plt.ylabel('DFT Phase |X(freq)| np.fft ')

    plt.figure(figsize = (20, 10))
    plt.subplot(121)
    plt.stem(freq, np.abs(X), 'b', markerfmt=" ", basefmt="-b")
    plt.xlabel('Freq (Hz)')
    plt.ylabel('DFT Amplitude |X(freq)|')
    plt.ylim(0, 6)

    plt.figure(figsize = (20, 10))
    plt.subplot(122)
    plt.stem(freq, np.angle(X), 'b', markerfmt=" ", basefmt="-b")
    plt.xlabel('Freq (Hz)')
    plt.ylabel('DFT Phase |X(freq)|')
    
    a = sum(abs(i)**2)
    b = 1/N*sum(abs(X)**2)
    c = 1/N*sum(abs(X1)**2)
    print(a, b, c)

plt.show()
```
